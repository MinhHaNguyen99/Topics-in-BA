```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Filter data for popularity > 70
filtered_data = data[data['popularity'] > 70]

# Step 2: Identify the top genres (from your earlier analysis)
top_genres = ['pop', 'reggaeton', 'reggae', 'latino', 'latin']

# Step 3: Filter data for top genres
filtered_data_top_genres = filtered_data[filtered_data['track_genre'].isin(top_genres)]

# Step 4: Select numerical columns
feature_numerical = [feature for feature in data.columns if data[feature].dtype != 'O']

# Step 5: Compute the correlation matrix for filtered data
correlation_matrix = filtered_data_top_genres[feature_numerical].corr()

# Step 6: Plot the correlation heatmap
plt.figure(figsize=(20, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation of Numerical Features for Popularity > 70 in Top Genres")
plt.show()
```

```python
from scipy.stats import pearsonr
```

```python
import pandas as pd

# List of numerical features excluding 'popularity' (independent variables)
independent_variables = [feature for feature in feature_numerical if feature != 'popularity']

# Calculate p-values for each independent variable with 'popularity'
p_values = {}
for feature in independent_variables:
    # Ensure no missing values in the filtered data
    clean_data = filtered_data_top_genres[['popularity', feature]].dropna()
    correlation, p_value = pearsonr(clean_data['popularity'], clean_data[feature])
    p_values[feature] = p_value

# Convert p-values to a DataFrame for easier visualization
p_values_df = pd.DataFrame(list(p_values.items()), columns=['Variable', 'P-Value'])
p_values_df.sort_values(by='P-Value', inplace=True)
```
